### Ejercicio 1

Cálculo de la probabilidad de obtener dos veces 6 al lanzar dos dados

```python
import numpy as np

def prob_x(dado1,dado2,dado3,x):
    suma = []
    for i in range(len(dado1)):
        suma.append(dado1[i] + dado2[i] +dado3[i])
    return [suma.count(x)/len(dado1), suma]

N = [n for n in range(10000, 100001, 10000)]
valor = 18
prob = []
for i in N:
    dado1 = np.random.randint(1,11,i)
    dado2 = np.random.randint(1,11,i)
    dado3 = np.random.randint(1,11,i)
    prob.append(prob_x(dado1,dado2, dado3,valor)[0])

for i in range(len(N)):
    print('Lanzando ' + str(N[i]) + ' veces 3 dados, la probabilidad de sumar ' + 
          str(valor) + ' es: ' +  str(prob[i]))
```

```python
suma = prob_x(dado1,dado2, dado3,valor)[1]
frec = [suma.count(i) for i in range(30)]
distr = [frec[i]/sum(frec) for i in range(len(frec))]
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(distr)
mas_probable = [i  for i in range(len(frec)) if frec[i] == max(frec)]
print('Más probable: ' + str(mas_probable))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from time import time
# integral del seno entre 0 y pi

#grafico el seno para verlo
eje_x = np.arange(0,np.pi,.01)
seno = [np.sin(eje_x[i]) for i in range(len(eje_x))]
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
ax1.plot(eje_x,seno, 'r')


# calculo integral con método de éxito fracaso
N = 1e5 # número de puntos

start = time()
ex_x = [] # posición de éxitos
ex_y = []
fra_x = []# posición de fracasos
fra_y = []
for i in range(int(N)):
    x = np.pi*np.random.rand()
    y = np.random.rand()
    if y <= np.sin(x):
        ex_x.append(x)
        ex_y.append(y)
    else:
        fra_x.append(x)
        fra_y.append(y)
        
exitos = len(ex_x)
integral = np.pi*exitos/N
tiempo_exfra = time() - start
texto_exito_fracaso = str(int(N)) + 'pasos \n integral = ' + str(integral) + '\n En ' + str(tiempo_exfra) + ' segundos'
ax1.scatter(ex_x,ex_y, s=1)
ax1.set_xlabel(texto_exito_fracaso)
ax1.set_title('Método del éxito-fracaso', fontsize=15)
ax1.scatter(fra_x,fra_y, c='g', s=1)


# calculo integral por método de media muestral
start = time()
valores = np.pi*np.random.rand(int(N))
funcion_valuada = [np.sin(valores[i]) for i in range(len(valores))]
sumatoria = sum(funcion_valuada)
media_muestral = sumatoria/int(N)

inte = np.pi*media_muestral
fin = time() - start
ax2.plot(eje_x,seno, 'r')
ax2.scatter(valores, funcion_valuada, c='g', s = 6)
ax2.set_title('Método de la media muestral', fontsize=15)
ax2.set_xlabel('Para ' + str(int(N)) + 'muestras \n la integral es: ' + str(inte) + '\n En ' + str(fin) + ' segundos')
plt.show()
```
